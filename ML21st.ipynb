## Q4.

```python
import pandas as pd
```

```python
df = pd.DataFrame({'Color': ('red', 'green', 'blue'),'Size': ('small', 'medium',
'large'), 'Material': ('wood', 'metal', 'plastic')})
```

```python
df
```

```python
## label encoding for color . for material and size its better to used ordinal.
```

```python
from sklearn.preprocessing import LabelEncoder

lbl_encoder=LabelEncoder()

encoded = lbl_encoder.fit_transform(df[['Color']]) ## one feature at a time . 

encoder_df=pd.DataFrame(encoded,columns=['color_label'])
```

```python
encoder_df
```

```python
## now for material and size .
```

```python
## ORdinal Encoding
from sklearn.preprocessing import OrdinalEncoder

## create an instance of ORdinalEncoder and then fit_transform
encoder=OrdinalEncoder(categories=[['small','medium','large']])

encoded = encoder.fit_transform(df[['Size']])

encoder_df_2=pd.DataFrame(encoded,columns=['Size_ordinal'])
```

```python
from sklearn.preprocessing import OrdinalEncoder

## create an instance of ORdinalEncoder and then fit_transform
encoder=OrdinalEncoder(categories=[['plastic','wood','metal']])

encoded = encoder.fit_transform(df[['Material']])

encoder_df_3=pd.DataFrame(encoded,columns=['Material_ordinal'])
```

```python
pd.concat([df,encoder_df,encoder_df_2,encoder_df_3],axis=1)
```
