Pyplot
Pyplot is a Matplotlib module which provides a MATLAB-like interface. Matplotlib is designed to be as usable as MATLAB, with the ability to use Python and the advantage of being free and open-source. Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, decorates the plot with labels, etc. The various plots we can utilize using Pyplot are Line Plot, Histogram, Scatter, 3D Plot, Image, Contour, and Polar.

## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as mp
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
mp.xlabel('x-axis')
mp.ylabel('y-axis')
mp.title('Scatter Plot')

mp.scatter(x,y)

mp.show()
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:


```python

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])

mp.subplot(2,3,1)
mp.plot(x,y)


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])

mp.subplot(2,3,2)
mp.plot(x,y)


x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
mp.subplot(2,3,4)
mp.plot(x,y)

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
mp.subplot(2,3,5)
mp.plot(x,y)
```

## Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.


```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
## bar plot 

mp.bar(company,profit)
mp.show()
```

```python
## horizontal bar plot 

mp.barh(company,profit)
```
