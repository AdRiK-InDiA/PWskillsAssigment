### Q 2. Road Tax

```python
cost_price = int(input('Enter to Cost Price of the Bike : '))

if cost_price>100000:
  print(f'Road Tax = {0.15*cost_price}')

elif cost_price>50000:
   print(f'Road Tax = {0.10*cost_price}')

else:
  print(f'Road Tax = {0.05*cost_price}')
```

### Q 5. When and where to use while loop ?

Python while loop is used to run a block code until a certain condition is met.

The syntax of while loop is:

    while condition:
          body of while loop

Here,

A while loop evaluates the condition
If the condition evaluates to True, the code inside the while loop is executed.
condition is evaluated again.
This process continues until the condition is False.
When condition evaluates to False, the loop stops.

With the while loop we can execute a set of statements as long as a condition is true.
The while loop requires relevant variables to be ready, in this example we need to define an indexing variable, i, which we set to 1.

**Pattern No 2**

```python
# python3 code to print pyramid pattern using while loop
n=5;i=0
while(i<=n):
  print(" " * (n - i) +"*" * i)
  i+=1
```

Pattern No.3

```python
n = int(input())

k = n - 1

for i in range(0, n):
  for j in range(0, k):
    print(end=" ")
  k = k - 1
  for j in range(0, i+1):
    print("* ", end="")
  print("\r")
```

### Q 7 . Reverse for loop

```python
n = int(input())

for i in range(n,0,-1):
  print(i,' ',end='')
```

###Q 8 . Reverse While Loop

```python
n = int(input())

while n>0:
  print(n,' ',end='')
  n-=1
```
